```python
import pandas as pd

base = '/home/jovyan/code/airbnb_data/japan/'

listings_file = base + 'listings.csv'

listings_df = df = pd.read_csv(listings_file)
```

```python
listings_req_cols = ['id','listing_url','name','description','picture_url','host_id','neighbourhood_cleansed',
                     'latitude',
       'longitude', 'property_type', 'room_type', 'accommodates',
       'bathrooms_text', 'bedrooms', 'beds', 'amenities', 'price',

                     'license', 'instant_bookable','reviews_per_month']
df_l = listings_df.filter(listings_req_cols)
```

```python

def show_na_cols(df,req_cols):
    na_cols = df.isnull().any(axis=0)
    na_cols_dict = dict(na_cols)
    for i in req_cols:
        if na_cols_dict[i]:
            print(i)  
#show_na_cols(df_l,listings_req_cols)
print('Before :',df_l.isnull().sum().sum())

listing_na_dict =  {'bathrooms_text':'1 bath','bedrooms':1,'beds':1,'reviews_per_month':0}
df_l = df_l.fillna(listing_na_dict)
print('After :',df_l.isnull().sum().sum())
```

```python
import re
def clean_text(df,cols):
    for c in cols:
        df[c] = df[c].apply(lambda text: re.sub(r'<[^>]*>|[\r\n]', '', text))
    return df

df_l['price'] = df_l['price'].apply(lambda text: re.sub(r'$|,', '', text)).str.lstrip('$').astype(float)
df_l['instant_bookable'] = df_l['instant_bookable'].astype(bool)
df_l = clean_text(df_l,['description'])

df_l.head(5)
```
