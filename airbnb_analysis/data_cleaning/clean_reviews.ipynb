```python
import pandas as pd

base = '/home/jovyan/code/airbnb_data/japan/'

reviews_file = base + 'reviews.csv'

df = reviews_pd = pd.read_csv(reviews_file)
reviews_pd.head(5)
```

```python
df = df.fillna('NA')
df.isnull().sum()
df.drop_duplicates()
```

```python
import re 
def clean_text(text):
    clean = re.sub(r'<[^>]*>|[\r\n]', '', text)
    return clean
df['comments']=  df['comments'].apply(clean_text)
df.head(5)
```

```python
df.to_csv(base+'/clean_data/reviews.csv',index=False,mode='w')
```
