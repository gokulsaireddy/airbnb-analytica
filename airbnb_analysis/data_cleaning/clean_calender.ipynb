```python
import pandas as pd

base = '/home/jovyan/code/airbnb_data/japan/'

calendar_file = base + 'calendar.csv'

calendar_df = df = pd.read_csv(calendar_file)
```

```python
import re 

df['date'] = df['date'].astype('datetime64[ns]')
for c in ['price','adjusted_price']:
    df[c] = df[c].apply(lambda text: re.sub(r'$|,', '', text)).str.lstrip('$').astype(float)
df['available'] = df['available'].astype(bool)
df.head(5)
```

```python
df.filter(['listing_id','price']).groupby(['listing_id']).aggregate(['max','min'])
```

```python
df.to_csv(base+'/clean_data/calendar.csv',index=False,mode='w')
```
